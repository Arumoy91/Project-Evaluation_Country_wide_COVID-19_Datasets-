```python
from sklearn.preprocessing import LabelEncoder

def object_to_int(dataframe_series):
    if dataframe_series.dtype=='object':
        dataframe_series = LabelEncoder().fit_transform(dataframe_series)
    return dataframe_series
```

```python
df = df.apply(lambda x: object_to_int(x))
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
X=df.drop(['Total Deaths'], axis=1)
y=df['Total Deaths']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 1)
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
df = DecisionTreeRegressor()
```

```python
df.fit(X_train,y_train)
```

```python
df.score(X_test,y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
lm.score(X_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
ad=AdaBoostRegressor()
```

```python
ad.fit(X_train,y_train)
```

```python
ad.score(X_test,y_test)
```

```python
gd=GradientBoostingRegressor()
```

```python
gd.fit(X_train,y_train)
```

```python
gd.score(X_test,y_test)
```

Conclusion: We have used several regression techniques for this dataset to find out total deaths due to covid19 virus and find out that RandomForestRegressor is the best suitable model for predicting the outcome of the dataset.
